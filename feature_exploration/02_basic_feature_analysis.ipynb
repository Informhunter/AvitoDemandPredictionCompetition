```python
import pandas as pd
import numpy as np
from zipfile import ZipFile
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import seaborn as sb
```

```python
sb.set_style('white')
sb.set(font_scale=2)
```

```python
with ZipFile('../data/train.csv.zip') as z:
    with z.open('train.csv') as f:
        train_df = pd.read_csv(f)
```

# Description

Minor feature exploration and extraction.

### region

```python
plt.figure(figsize=(15, 7))
sb.boxplot(train_df.loc[train_df['deal_probability'] != 0, 'region'],
           train_df.loc[train_df['deal_probability'] != 0, 'deal_probability'])
plt.xticks(rotation=90);
```

### parent_category_name

```python
plt.figure(figsize=(10, 5))
sb.boxplot(train_df.loc[train_df['deal_probability'] != 0, 'parent_category_name'],
           train_df.loc[train_df['deal_probability'] != 0, 'deal_probability'])
plt.xticks(rotation=90);
```

### category_name

```python
plt.figure(figsize=(15, 7))
sb.boxplot(train_df.loc[train_df['deal_probability'] != 0, 'category_name'],
           train_df.loc[train_df['deal_probability'] != 0, 'deal_probability'])
plt.xticks(rotation=90);
```

### price

```python
train_df['has_price'] = ~pd.isnull(train_df['price'])
```

```python
sb.boxplot(train_df['has_price'], train_df.loc[train_df['deal_probability'] != 0, 'deal_probability'])
```

### item_seq_number

```python
sb.jointplot(train_df.loc[train_df['deal_probability'] != 0, 'item_seq_number'],
             train_df.loc[train_df['deal_probability'] != 0, 'deal_probability'])
```

```python
sb.jointplot(train_df.loc[np.logical_and(train_df['deal_probability'] != 0, train_df['item_seq_number'] < 100), 'item_seq_number'],
             train_df.loc[np.logical_and(train_df['deal_probability'] != 0, train_df['item_seq_number'] < 100), 'deal_probability'])
```

### user_type

```python
sb.boxplot(train_df.loc[train_df['deal_probability'] != 0, 'user_type'],
           train_df.loc[train_df['deal_probability'] != 0, 'deal_probability'])
```

### image

### image_top1

```python
sb.jointplot(train_df.loc[train_df['deal_probability'] != 0, 'image_top_1'],
             train_df.loc[train_df['deal_probability'] != 0, 'deal_probability'])
```

### activation_date

```python
train_df['activation_date'] = pd.to_datetime(train_df['activation_date'])
```

```python
train_df['activation_dayofweek'] = train_df['activation_date'].dt.dayofweek
```

```python
sb.boxplot(train_df.loc[train_df['deal_probability'] != 0, 'activation_dayofweek'],
           train_df.loc[train_df['deal_probability'] != 0, 'deal_probability'])
```
